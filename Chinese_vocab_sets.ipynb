# Animate SUBJ/OBJ

```python
indicatives = np.array(['有个','这个','那个', '有些','这些','那些'])

def subjGenerator(index):
    #np.random.seed(0)
    Subj = vocab['expression_zh'][index]
    if vocab['properNoun'][index] ==1 and vocab['animate'][index] ==1:
        return Subj
    elif (vocab['properNoun'][index] == 0) and (vocab['animate'][index] ==1) and vocab['category'][index] != 'N/NP' and vocab['category'][index] != 'N\\NP[poss]':
        return np.random.choice(indicatives) + Subj
    elif vocab['category'][index] == 'N\\NP[poss]':
        Proper_Subj = np.random.choice(all_animate_proper_nouns) + '的' + Subj
        Common_Subj = np.random.choice(indicatives) + np.random.choice(all_singular_animate_common_nouns) + '的' + Subj
        return np.random.choice([Proper_Subj,Common_Subj])
    elif (vocab['category'][index] == 'N/NP') and (vocab['category_2'][index] == 'RelN'):
        return np.random.choice(all_animate_proper_nouns) + Subj
    else:
        return Subj
```

# Classifier

```python
def classifier(num):
    np.random.seed(0)
    c = 0
    lst = []
    dic = {}
    while c < num:
        
        TV_index = np.random.choice(all_transitive_verbs.index)
        TV = all_transitive_verbs[TV_index]
        
        Arg1_index = np.random.choice(vocab['expression_zh'].loc[eval(vocab['arg_1_zh'][TV_index])].index)
        Arg1 = subjGenerator(Arg1_index)        
        Arg2_index = np.random.choice(all_singular_nouns.loc[eval(vocab['arg_2_zh'][TV_index])].index)
        Arg2 = vocab['expression_zh'][Arg2_index]
                
        Classifier1 = vocab['classifier_type'][Arg2_index]
        if pd.isna(Classifier1) == True:
            continue
        if Classifier1 not in dic:
            dic[Classifier1] = 0
            dic[Classifier1]+=1
        else:
            if dic[Classifier1] == 50:
                continue
            else:
                dic[Classifier1] += 1

        Classifier0 = np.random.choice(all_classifiers[(all_classifiers != Classifier1)& (all_classifiers !='次') & (all_classifiers !='位')& (all_classifiers !='个')])
        
        Sentence1 = Arg1 + TV + '一' + Classifier1 + Arg2
        Sentence0 = Arg1 + TV + '一' + Classifier0 + Arg2
        if ['classifier','classifier',Sentence1,1] and ['classifier','classifier',Sentence0,0] not in lst:
            lst.append(['classifier','classifier',Sentence1,1])
            lst.append(['classifier','classifier',Sentence0,0])
        c += 1
    return lst,dic
```

```python
def evenGenerator(num, width):
    np.random.seed(1)
    lst = []
    c = 0
    while c < num:
        n = np.random.randint(0, width)
        if n%2 == 0:
            if n not in lst:
                lst.append(n)
                c += 1
        else:
            pass
    return lst
```

```python
shuffle_row_index = evenGenerator(45,180)
```

```python
def shuffle(df,index):
    for num in index:
        a = df.loc[num]
        b = df.loc[num+1]
        df.loc[num] = b.values
        df.loc[num+1] = a.values
    return df
```
